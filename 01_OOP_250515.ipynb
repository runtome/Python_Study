```python
class ClassName:
    def __init__(self, arg1, arg2):       # Constructor method
        self.arg1 = arg1                  # Instance attribute
        self.arg2 = arg2

    def _private_method(self):            # Instance method
        return self.arg1

    def public_method(self, x):
        '''
        Example of use
        input name = str

        '''
                 # Method with extra argument
        return self.arg2 + x
```

```python
class PlayerCharacter:
    def __init__(self, name, age): #dunder method , self refer to itself
        self.name = name
        self.age = age

    def run(self): # Instance method
        print("run")

    def len(self): # Instance method
        return len(self.name)
```

```python
# Create instance
player1 = PlayerCharacter()
print(player1)
```

```python
player1 = PlayerCharacter(age= 22 , name = "Sam" )
print(player1.name)
print(player1.age)
player1.run()
```

```python
player2 =PlayerCharacter("Tomwwww|" , 34)
```

```python
player1.age+player2.age
```

```python
print(player1.len())
```

```python
dir(player1)
```

```python
len(player2)
```

```python
print(player1)
print(player2)
```

```python
#if only referobject
# Create an object
player1 = PlayerCharacter("Alice", 100)

# Assign player2 to refer to the same object as player1
player2 = player1

# Modify using player2
player2.age = 50

print("player1:", player1.age)
print("player2:", player2.age)

# Print both to show that both reflect the change
print("player1:", player1)
print("player2:", player2)
```
